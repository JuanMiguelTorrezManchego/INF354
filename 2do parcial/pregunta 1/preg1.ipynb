```python

from sklearn.datasets import load_iris
import numpy as np
import random
dato=load_iris()
X1=dato.data
y1=dato.target

ran = [np.random.randint(0,150) for i in range(100)]
#print(ran)
X =[]
y=[]
for i in range(100):
  X.append(X1[ran[i]])
  y.append(y1[ran[i]])

#print(X)
#print(y)

#print(X)
print(y)
pesos = [[np.random.uniform(low=0, high= 1) for j in range(4)] for i in range(100)]
bias = np.array([0.0])
#print(pesos)

def sigmoid(x):
  return (1/(1+np.exp(-x)))

def tanHip(x):
  return ((1-np.exp(-2*x))/(1+np.exp(-2*x)))

def ReLu(x):
  return np.max(x,0)


def prediccion(entradas_vector, pesos_vector, bias_vector):
  capa1 = np.dot(entradas_vector, pesos_vector) + bias_vector
  capa2 = ReLu(capa1)
  return capa2

y_esp=[]
for i in range(100):
  #print(X[i]," ////" ,pesos[i])
  aux = prediccion(X[i], pesos[i], bias)
  y_esp.append(aux)

#print(y_esp)
y_esp1=[]
for i in range(100):
  if(y_esp[i]<6): y_esp1.append(0)
  elif (y_esp[i]<8): y_esp1.append(1)
  else : y_esp1.append(2)



print( y_esp1)

#error cuadratico medio
def error(ypredic, yverd):
  er= 0
  for i in range(100):
    er += np.abs(ypredic[i]-yverd[i])
    #print((ypredic[i]-yverd[i])*(ypredic[i]-yverd[i]))
  #print(er)
  er = er/100
  return er

print ("error cuadratico medio:", error(y_esp1,y))
```
